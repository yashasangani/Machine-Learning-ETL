```python
def replacer(df):
    cat = list(df.columns[df.dtypes == "object"])
    for i in df.columns:
        if i in cat:
            md = df[i].mode()[0]
            df[i] = df[i].fillna(md)
        else:
            means= df[i].mean()
            df[i] = df[i].fillna(means)

    
```

### Feature Engineering

1. cat - Onehotencoding is preferred

```python
from sklearn.preprocessing import OneHotEncoder
x_cat = x[cat]
x_con = x[con]
```

```python
x_cat
```

```python
x_cat["Cylinders"].unique()
```

```python

ohe = OneHotEncoder(
    handle_unknown='ignore',
    sparse_output=False
    ).set_output(transform="pandas")
x_cat_pre = ohe.fit_transform(x_cat)
```

```python
x_cat_pre
```

```python
x_cat_pre.to_csv("categorical.csv",index=False)
```

### Apply standard scaler on continous (numerical )

```python
x_con
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().set_output(transform="pandas")
x_con_pre = scaler.fit_transform(x_con)
x_con_pre
```

### Join x_cat_pre and x_con_pre

```python
x_pre = x_con_pre.join(x_cat_pre)
x_pre.head()
```

```python
x_pre.to_csv("Preprocessed.csv",index=False)
```

### Train Test split

Test size should be between 20 and 30

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(x_pre,y,test_size=0.2,random_state=42)
```

```python
ytrain
```

### Fit the model on train and test

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain,ytrain)
```

### Model evaluation

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error,r2_score
```

```python
def evaluate_model(model,x,y):
    #predict the result
    ypred = model.predict(x)
    #evaluate the metrics
    mse = mean_squared_error(y,ypred)
    rmse=mse**(1/2)
    mae = mean_absolute_error(y,ypred)
    mape = mean_absolute_percentage_error(y,ypred)
    r2 = r2_score(y,ypred)
    #print the result
    print(f"MSE : {mse:.2f}")
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.4f}")
    print(f"R2Score : {r2:.4f}")

```

```python
evaluate_model(model,xtrain,ytrain)
```

```python
ypred_train = model.predict(xtrain)
ypred_train[0:5]
```

```python
ytrain[0:5]
```

```python
evaluate_model(model,xtest,ytest)
```

```python
ypred_ytest = model.predict(xtest)
ypred_ytest[0:5].round(3)
```
